```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

## **Считывание данных**

```python
train = pd.read_csv("train.csv")
formula_train = pd.read_csv("formula_train.csv")

print(f"Размер датасета train: {train.shape}")
print(f"Размер датасета formula_train: {formula_train.shape}")
```

```python
test = pd.read_csv("test.csv")
formula_test = pd.read_csv("formula_test.csv")

print(f"Размер датасета test: {test.shape}")
print(f"Размер датасета formula_test: {formula_test.shape}")
```

## **Обработка данных и формирование датасета**

```python
print("Число пропущенных значений в тренировочном датасете:")
print(train.isna().sum().sum())
print(formula_train.isna().sum().sum())
```

```python
print("Число пропущенных значений в тестовом датасете:")
print(test.isna().sum().sum())
print(formula_test.isna().sum().sum())
```

```python
formula_train = formula_train.drop(columns=['critical_temp'])
train_full = pd.concat([train, formula_train], axis=1)
print(f"Размер полного тренировочного датасета: {train_full.shape}")
```

```python
train_full.drop(columns=['material'], inplace=True)
```

```python
x = train_full.drop(columns=['critical_temp'])
y = train_full['critical_temp']

print(f"Размер вектора признаков: {x.shape}")
print(f"Размер вектора ответов: {y.shape}")
```

```python
test_full = pd.concat([test, formula_test], axis=1)
print(f"Размер полного тестового датасета: {test_full.shape}")
```

```python
test_full.drop(columns=['material'], inplace=True)
```

## **Обучение модели с помощью метода отложенной выборки**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
print(f"Размер тренировочного датасета: {x_train.shape}, {y_train.shape}")
print(f"Размер тестового датасета: {x_test.shape}, {y_test.shape}")
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
print('Веса всех признаков (w1, ..., w167): ', model.coef_)
print('Свободный коэффицент уравнения w0:', model.intercept_)
```

```python
features = test_full.columns

coeff_df = pd.DataFrame(model.coef_, columns=['Coefficient'])
coeff_df['features'] = features

coeff_df.sort_values(by='Coefficient')
```

```python
y_pred = model.predict(x_test)

print("Среднеквадратичная ошибка: ", mean_squared_error(y_test, y_pred))
```
